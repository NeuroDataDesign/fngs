```python
%matplotlib inline
import matplotlib
import numpy as np
import matplotlib.pyplot as plt

x = [0, 1]
xlabels = ['BNU1', 'LMU1']

orig = [0.9336, 0.7852]
epireg = [0.9100, 0.7604]
mniepiflirt = [0.9210, 0.7664]
mniepiflirtfnirt0 = [0.8916, None]
mniepiflirtfnirt1 = [0.7046, None]
mniepiflirtfnirt2 = [0.7112, None]

plt.scatter(x, orig, color='r', label="Original")
plt.scatter(x, epireg, color='b', label ="EPIREG")
plt.scatter(x, mniepiflirt, color='y', label="MNI-EPI-Template + FLIRT")
plt.scatter(x, mniepiflirtfnirt0, label="MNI-EPI-Template + FLIRT + FNIRT (No tweaks)")
plt.scatter(x, mniepiflirtfnirt1, label="MNI-EPI-Template + FLIRT + FNIRT (MNI152 Config)")
plt.scatter(x, mniepiflirtfnirt2, label="MNI-EPI-Template + FLIRT + FNIRT (MNI152 Config + Membrane Energy)")


plt.ylabel("MNR")
plt.xlabel("Dataset")
plt.legend(loc='center', bbox_to_anchor=(1.8,0.5))
plt.title("Discriminability Scores")
plt.xticks(x, xlabels)

plt.show()
```
